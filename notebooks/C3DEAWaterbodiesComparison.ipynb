# C3 DEA Waterbodies comparison

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
C3 = pd.read_csv('AusWaterBodiesFilteredx33_y23.csv')
v1 = pd.read_csv('v1 Kati Thanda South.csv')
```

```python
plt.figure(figsize=[10,5])
plt.hist([C3['area'] / 1000000, v1['area']/ 1000000], bins=20, range=(0, 4.2), 
         color = ['k', 'r'], log=True, label=[f'Test n=({len(C3)})', f'DEA Waterbodies v1 n=({len(v1)})']);
plt.legend(loc='upper right');
ax = plt.gca()
ax.get_xaxis().get_major_formatter().set_scientific(False)
plt.xlabel('Waterbody area (km$^2$)');
plt.ylabel('Histogram count');
plt.title('Kati Thanda South, SA test tile');
```
